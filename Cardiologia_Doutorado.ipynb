DICIONÁRIO DOS DADOS:


1-**Idade** - age: idade em anos.

2-**Sexo**: sexo do paciente

0 = feminino;

1 = masculino.

3-**Dor_Peito**: Cp (Chest Pain Type - Cp)- tipo de dor no peito experimentada pelo paciente, onde:

1: Ausência de dor no peito.

2: Dor não anginosa.

3: Dor anginosa típica.

4: Dor anginosa atípica.

4-**Pressao_Arterial** em Repouso (Resting Blood Pressure - Trestbps): Trestbps - pressão arterial em repouso (em mm Hg).

5-**Colesterol_Serico** (Serum Cholesterol - fbs): Col - colesterol sérico em mg/dl.

6-**Acucar_Maior_120** em Jejum (Fasting Blood Sugar): fbs - níveis de açúcar no sangue em jejum > 120 mg/dl .

1 = verdadeiro;

0 = falso.

7-**ECG** Eletrocardiográfico em Repouso (Resting Electrocardiographic Results - restecg): **resultados do eletrocardiograma** em repouso, onde:

0: Normal

1: Anormalidades de ST-T (inversões T e/ou elevação ou depressão de > 0.05 mV)

2: Hipertrofia ventricular esquerda provável ou definitiva (pelo critério de Estes)

8-**Freq_Cardiaca**: thalach - Frequência Cardíaca Máxima Alcançada (Maximum Heart Rate Achieved - thalach).

9-**Ex_Angina**: exang - Angina Induzida por Exercício (Exercise Induced Angina - exang)

1 = sim;

0 = não

10-**Dep_Seg_ST** - Depressão do Segmento ST Induzida por Exercício (ST Depression Induced by Exercise - oldpeak)

11-**Inc_Seg_ST** : Slope - Inclinação do Segmento ST Máxima (Peak Exercise ST Slope) durante o exercício, onde:

0: Inclinação ascendente.

1: Plana.

2: Inclinação descendente.

12-**VasosP_Coloridos**:Ca - número de vasos principais coloridos por fluoroscopia:

0

1

2

3

13-thal - **Talassemia** (Thalassemia - thal): uma variante da anemia talassêmica

3 = normal;

6 = defeito fixo;

7 = defeito reversível.

14-**Grav_Doenca_cardiaca** - num - Portanto, a coluna 'num' é a variável de destino, que indica a presença de doença cardíaca. Os valores podem ser:

0: Ausência de doença cardíaca

1: Presença de uma doença cardíaca (1 vaso colorido)

2: Presença de uma doença cardíaca (2 vasos coloridos)

3: Presença de uma doença cardíaca (3 vasos coloridos)

4: Presença de uma doença cardíaca (4 vasos coloridos)

Em resumo, a coluna 'num' é a variável que você tentará prever usando as outras características do conjunto de dados.
A coluna 'num' se refere ao número de vasos principais coloridos por fluoroscopia e pode conter os seguintes valores:
0: Ausência de doença cardíaca
1: Presença de uma doença cardíaca (1 vaso colorido)
2: Presença de uma doença cardíaca (2 vasos coloridos)
3: Presença de uma doença cardíaca (3 vasos coloridos)
4: Presença de uma doença cardíaca (4 vasos coloridos)
Esses valores indicam a gravidade da doença cardíaca, com um número maior de vasos coloridos sugerindo uma doença mais grave. Portanto, a coluna 'num' serve como uma variável de destino que indica a gravidade da doença cardíaca em vez de simplesmente sua presença ou ausência.

# **1. Modelos - Tentando prever as diversas classes de doenças (0 a 4)**

As técnicas **Random Forest**, **XGBoost** e **Decision Trees** pertencem à mesma família de algoritmos de aprendizado de máquina baseados em árvores de decisão. Esses métodos compartilham a característica fundamental de utilizar estruturas de árvore para realizar previsões e classificação. As Árvores de Decisão (Decision Trees) são a base dessa família, sendo algoritmos simples que segmentam os dados em grupos sucessivamente menores através de decisões baseadas em atributos. O Random Forest expande essa ideia criando múltiplas árvores de decisão em paralelo e combinando suas previsões para aumentar a robustez e reduzir o overfitting. O **XGBoost** (Extreme Gradient Boosting) avança ainda mais ao aplicar uma técnica de boosting, onde as árvores são construídas sequencialmente, com cada nova árvore corrigindo os erros das anteriores, resultando em um modelo altamente preciso e eficiente.

As árvores de decisão, Random Forest e XGBoost são todos algoritmos de aprendizado de máquina usados principalmente para tarefas de classificação e regressão. Embora estejam relacionados, eles têm diferenças significativas em termos de estrutura, funcionamento e aplicação. Aqui está um resumo das principais diferenças entre eles:

Árvore de Decisão

Estrutura: Uma árvore de decisão é um modelo preditivo que divide os dados em subconjuntos menores com base em características específicas dos dados. Cada nó interno representa uma "pergunta" sobre um atributo, cada ramo representa o resultado dessa pergunta, e cada folha representa uma classe ou valor de regressão.

Funcionamento: As árvores de decisão funcionam criando divisões binárias nos dados até que cada partição contenha amostras predominantemente de uma única classe (ou até um critério de parada seja atingido).

Vantagens: Simplicidade, interpretabilidade e fácil visualização. São rápidas para treinar.

Desvantagens: Tendem a sobreajustar (overfitting) aos dados de treinamento, o que pode levar a uma menor generalização.

Random Forest

Estrutura: Random Forest é um conjunto (ensemble) de várias árvores de decisão, onde cada árvore é construída a partir de um subconjunto aleatório dos dados de treinamento e das características.

Funcionamento: A decisão final é tomada por votação majoritária no caso de classificação, ou pela média das previsões no caso de regressão. Cada árvore no conjunto é construída de forma independente.

Vantagens: Reduz o problema de sobreajuste das árvores de decisão individuais, melhorando a precisão e a robustez do modelo.

Desvantagens: Menos interpretável do que uma única árvore de decisão, mais complexo e computacionalmente intensivo.

XGBoost (Extreme Gradient Boosting)

Estrutura: XGBoost é um algoritmo de boosting, uma técnica de ensemble que combina preditores fracos para formar um preditor forte. Ele constrói árvores de decisão sequencialmente, onde cada nova árvore corrige os erros das árvores anteriores.

Funcionamento: Utiliza uma abordagem de gradiente para minimizar uma função de perda, ajustando os pesos das observações com erros maiores. Além disso, implementa técnicas avançadas como regularização e manejo eficiente de valores ausentes.

Vantagens: Alta precisão, eficiente em termos de tempo de treinamento e uso de memória, capacidade de lidar com dados esparsos e com valores ausentes, e controle de sobreajuste através de regularização.

Desvantagens: Pode ser mais difícil de interpretar, maior complexidade e necessidade de ajuste cuidadoso dos hiperparâmetros.
Resumo das Diferenças
Complexidade e Precisão:

Árvore de Decisão: Simples, pode sobreajustar.
Random Forest: Mais complexo que uma única árvore, menos propenso a sobreajuste.
XGBoost: Altamente complexo e preciso, lida bem com sobreajuste.
Tempo de Treinamento:

Árvore de Decisão: Rápido.
Random Forest: Moderado, devido ao treinamento de múltiplas árvores.
XGBoost: Pode ser mais lento devido à natureza iterativa do algoritmo.
Interpretabilidade:

Árvore de Decisão: Alta.
Random Forest: Moderada, mais difícil de interpretar que uma única árvore.
XGBoost: Baixa, devido à complexidade do modelo.
